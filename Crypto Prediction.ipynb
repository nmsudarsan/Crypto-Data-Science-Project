## Random Forest using Grid Search Parameters

### Grid Search

```python
from sklearn.model_selection import GridSearchCV

# Split the data into features (X) and target variable (y)
X = tmp2.drop(['ID', 'Target'], axis=1)
y = tmp2['Target']

X_train = X
y_train = y
X_test = tmp3.drop(['ID', 'Target'], axis=1)
y_test = tmp3['Target']

# StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define RF parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Fit grid search on RF Classifier
rf = RandomForestClassifier(random_state=999)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy found: ", grid_search.best_score_)

# Use the best estimator to make predictions
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Evaluate the model
print("Test Accuracy: ", accuracy_score(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
f1_weighted = f1_score(y_test, y_pred, average='weighted')
print("Weighted Average F1 Score on Test Data with Best Model:", f1_weighted)
```

```python
# Split the data into features (X) and target variable (y)
X = tmp2.drop(['ID', 'Target'], axis=1)
y = tmp2['Target']

X_train = X
y_train = y
X_test = tmp3.drop(['ID', 'Target'], axis=1)
y_test = tmp3['Target']

# Setting random seed
rng = np.random.RandomState(999)
indices_train = rng.permutation(len(X_train))
indices_test = rng.permutation(len(X_test))

# Use the shuffled indices to shuffle the data
X_train = X_train.iloc[indices_train]
y_train = y_train.iloc[indices_train]
X_test = X_test.iloc[indices_test]
y_test = y_test.iloc[indices_test]

# Minmax scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize the Random Forest classifier
RF = RandomForestClassifier(
    n_estimators=300,    # Number of trees
    max_depth=20, # Max number of levels in each tree
    min_samples_split=2,  # Min number of samples to split a node
    min_samples_leaf=1,   # Min number of samples at each leaf node
    bootstrap=False, # Method of selecting samples for training each tree
    random_state=999
)

# Train the Random Forest classifier
RF.fit(X_train, y_train)

# Make predictions on the test set
y_pred = RF.predict(X_test)

# Predict on the training data
y_train_pred = RF.predict(X_train)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_pred)

# Print accuracies to compare
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
f1_weighted = f1_score(y_test, y_pred, average='weighted')
print("Weighted Average F1 Score on Test Data with Best Model:", f1_weighted)
```

- Precision measures how many of the predicted positives were actually positives. In our case our model was 74% effective at doing so. Precision also measure how many of the predicted negatives were actually negatives. In our case our model was 83% effective at doing so.
- Our accuracy was 77% meaning 77% of cases were correctly identified.

Weighted Average F1 Score: 0.75666796

Further explanation: Our model had higher recall than precision meaning that it was better at finding positives than rejecting false positives. Our model therefore casted a "wider net" and captured more positives than it should have but, in the process it found a fairly high percentage of the total positive cases. Precision is higher than recall for class 0.

Our model can be said to have been moderately successful but, above average in its prediction capabilities.

# Discussion/Conclusion

1. Our best model utilized a random forest approach with parameters from a grid search. This method built multiple decision trees and averaged their results to improve accuracy and reduce bias. The grid search explored different parameters to optimize performance.

2. Our team was successful in creating various different models including descision tree, random forrest, naive bayes, etc. However, fairly early on in our process we found that it was hard to break that 75% accuracy threshold and after many rounds of experimentation including trying to replace null values with different values and using oversampling and smote, we ended up settling with the RF model.

3. We found that utilizing the median values to replace null values was the most effective. This is most likely do to with there being so many null values for our features that just dropping the nulls wouldn't be as effective. Additionally, the median is better at avoiding the influence of outliers than other measures of central tendency like mean.

4. An early key insight that we found was that the target variable (whether the price of crypto went up or down) was fairly balanced over this period despite the overall rise in crypto price during 2023. We hypothesize this is due to rapid spikes due to the volatility of the crypto market which led to the events like the dramatic rise in crypto prices in late 2023. This volatitlity is apart of the nature of crypto due to its lack of usability in the present day and the major influence of hype and online media coverage on its price.

5. Our team tried out oversampling and SMOTE as a last ditch effort to try to increase the accuracy of our model. Despite these efforts, the realtively balanced nature of our dataset cause our models to produce similar results to the original split so we decided to stick with this.
